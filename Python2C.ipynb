```python
import xml.etree.ElementTree as ET
root = ET.parse('./tile.xml').getroot()

str="u16 walkableTiles[]={"
i=0

for tile in root.findall('tile'):
    id = tile.get('id')
   
    for prprts in tile.findall('./properties/property'):
        pptname = prprts.get('name')
        pptval = prprts.get('value')
       
        if pptname == "walkable" and pptval == "true":
            if i == 0:
                str = str + id
                i=i+1
            else:
                str = str + ", " + id
   
str = str + "};"

print(str)
```
